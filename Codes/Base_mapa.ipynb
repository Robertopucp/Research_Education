```r
base <- data.frame(read_excel("../Censo/Manzan_lima.xlsx"))
```

```r
dim(base)
```

```r
# slice data i tow. One IDMANZANA without MANZANA_A, second with MANZANA_A
data1 <- base %>% filter(str_sub(base$IDMANZANA,-1) == "0") #IDMANZANA without MANZANA_A
data2 <- base %>% filter(str_sub(base$IDMANZANA,-1) != "0") #IDMANZANA with MANZANA_A

# Drop 0 in the end of IDMANZANA
data1 <-data1 %>% mutate(IDMANZANA = str_sub(IDMANZANA,1,nchar(IDMANZANA)-1)) 

# Concat dataframes
data <- rbind(data1, data2)
```

```r
data
```

```r
# create id_manzana to match id_manzana from cartography base

data$id_manzana <- paste(str_sub(data$IDMANZANA,1,6), "0001", str_sub(data$IDMANZANA,7,nchar(data$IDMANZANA)), sep ="")
data <- data %>% select(id_manzana, everything())
data
```

```r
# selecting variables 

Fdata <- data %>% select(id_manzana,UBIGEO, CODCCPP,ZONA_ID,ZONA_A,MANZANA_ID,MANZANA_A,GRUPOS_QUIN_4,GRUPOS_QUIN_5,GRUPOS_QUIN_6,C5_P13_3,C5_P13_4,
                         POB_TOTAL,C5_P8_1,C5_P8_2,C5_P8_4,C5_P13_6,C5_P13_7,C5_P13_8,C5_P13_9,C5_P13_10,PET,PEA,PEAO,C5_P22_1,C5_P22_2,C5_P22_3,C5_P22_4,GRUPOS_EDAD_2,
                            GRUPOS_EDAD_3, GRUPOS_EDAD_4, GRUPOS_EDAD_5)

Fdata

# making each variable as an object

attach(Fdata)
```

```r
Fdata <- Fdata %>% rename(GRUPO1 = GRUPOS_QUIN_4, GRUPO2 = GRUPOS_QUIN_5, GRUPO3 = GRUPOS_QUIN_6)
```

```r
# var 1: age 15-25 / total population per each manzana
# var 2: age 15-25 / PET
# var 3: age 15-29 / total population per each manzana
# var 4: age 15-29 / PET
# var 5: PEA / PET
# var 6 : (PEA - PEAO) / PEA
# var 7: PEAO/PEA
# var 8: SIS/POB_TOTAL
# var 9: ESSALUD/POB_TOTAL
# var 10: Seguro privado /POB_TOTAL
# var11: n° educación primaria/ POB_TOTAL
# var12: n° educación secundaria/ POB_TOTAL
# var13: n° educación superior universitaria incompleta / POB_TOTAL
# var14: n° educación superior universitaria incompleta / PET
# var15: n° educación superior universitaria completa / POB_TOTAL
# var16: n° educación superior universitaria completa / PET
# var17: n° educación superior universitaria incompleta + completa /POB_TOTAL
# var18: n° educación superior universitaria incompleta + completa / PET


Fdata <- Fdata %>% mutate(var1 = (GRUPO1+GRUPO2)/POB_TOTAL, var2 = (GRUPO1+GRUPO2)/PET, 
                          var3 = (GRUPO1+GRUPO2+GRUPO3)/POB_TOTAL, var4 = (GRUPO1+GRUPO2+GRUPO3)/PET, 
                          var5 = PEAO / PET, var6 = (PEA - PEAO) / PEA, var7 = PEAO/PEA,
                          var8 = C5_P8_1/POB_TOTAL, var9 = C5_P8_2/POB_TOTAL, var10 = C5_P8_4/POB_TOTAL,
                          var11 = C5_P13_3/(POB_TOTAL), var12 = C5_P13_4/(POB_TOTAL),
                          var13 = C5_P13_8/(POB_TOTAL), var14 = C5_P13_8/(PET),
                          var15 = C5_P13_9/(POB_TOTAL), var16 = C5_P13_9/(PET),
                          var17 = (C5_P13_8+C5_P13_9)/(POB_TOTAL), var18 = (C5_P13_8+C5_P13_9)/(PET))


```

```r
#Cheking duplicates 
Fdata$duplicates <- duplicated(Fdata$id_manzana)

#No duplicates report 
Fdata %>% filter(duplicates == 1)
```

```r
#save data in .dta extension

write_dta(Fdata, '../Censo/data_manzana.dta')
```
